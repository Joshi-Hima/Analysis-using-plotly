```python
trace1 = go.Scatter(
    x = df.index,
    y = df.candidate_confidence,
    mode = "lines",
    name = "candidate_confidence",
    marker = dict(color = 'rgba(80,27,80,0.7)'),
    text = df.sentiment
)
data1 = [trace1]

layout = dict(title = "Candidate Confidence analysis",
             xaxis = dict(title = "index", ticklen = 5, zeroline = False))

fig = dict(data = data1, layout = layout)
iplot(fig)
```

# lets look at the people whose sentiment is positive, negative and neutral

```python
dfPositive = df[df['sentiment'] == 'Positive']
dfNegative = df[df['sentiment'] == 'Negative']
dfNeutral = df[df['sentiment'] == 'Neutral']

trace1 = go.Scatter(
    x = dfPositive.index,
    y = dfPositive.sentiment_confidence,
    name = "Positive People",
    mode = "markers",
    marker = dict(color = 'rgba(255, 128, 2, 0.8)'),
    text = dfPositive.candidate
)

trace2 = go.Scatter(
    x = dfNegative.index,
    y = dfNegative.sentiment_confidence,
    name = "Negative People",
    mode = "markers",
   marker = dict(color = 'rgba(255, 128, 200, 0.8)'),
    text = dfNegative.candidate
)

trace3 = go.Scatter(
    x = dfNeutral.index,
    y = dfNeutral.sentiment_confidence,
    name = "Neutral People",
    mode = "markers",
    marker = dict(color = 'rgba(255, 128, 255, 0.8)'),
    text = dfNeutral.candidate
)

data2 = [trace1, trace2, trace3]

layout = dict(title = "Visualization of Postive, Negative, and Neutral Poeple",
             xaxis = dict(title = 'index', ticklen = 5, zeroline = False),
             yaxis = dict(title = 'sentiment_confidence', ticklen = 5, zeroline = False)
             )

fig = dict(data = data2, layout = layout)
iplot(fig)
```

```python
df['tweet_created'] = pd.to_datetime(df['tweet_created'], utc=True)

# Extract the date component
df['date'] = df['tweet_created'].dt.date

# Extract the time component
df['time'] = df['tweet_created'].dt.time

# Extract the timezone component by splitting the original string (this will be a string operation)
df['timezone'] = df['tweet_created'].astype(str).str.split().str[-1]

# Display the modified DataFrame
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])

# Extract 'year'
df['year'] = df['date'].dt.year

print(df[['date', 'year']])
```

```python
sentiment_counts = df.groupby(['year', 'sentiment']).size().reset_index(name='count')

# Group by year to get average sentiment confidence
average_confidence = df.groupby('year')['sentiment_confidence'].mean().reset_index(name='average_confidence')

# Prepare data for plotting
sentiment_counts['year_sentiment'] = sentiment_counts['year'].astype(str) + ' ' + sentiment_counts['sentiment']
average_confidence['year'] = average_confidence['year'].astype(str) + ' Year'
```

```python
# Create bar traces
trace1 = go.Bar(
    x=sentiment_counts['year_sentiment'],
    y=sentiment_counts['count'],
    name='Tweet Count by Sentiment',
    marker=dict(color='rgba(255, 58, 255, 0.4)', line=dict(color='rgb(0,0,0)', width=1.5)),
    text=sentiment_counts['count']
)

trace2 = go.Bar(
    x=average_confidence['year'],
    y=average_confidence['average_confidence'],
    name='Average Sentiment Confidence',
    marker=dict(color='rgba(15, 15, 250, 0.4)', line=dict(color='rgb(1,1,1)', width=1.5)),
    text=average_confidence['average_confidence']
)

# Combine traces and define layout
data4 = [trace1, trace2]
layout = go.Layout(
    barmode='group',
    title='Sentiment Analysis Over Time',
    xaxis_title='Year / Sentiment',
    yaxis_title='Count / Average Confidence'
)

# Create the figure and plot it
fig = go.Figure(data=data4, layout=layout)
iplot(fig)
```

```python
data_sorted = df.sort_values(by='sentiment_confidence', ascending=True)
data = data_sorted.iloc[:20, :].copy()
data.index = range(0, len(data))

plot_data = [
    {
        'y': data.sentiment_confidence,
        'x': data.index,
        'mode': 'markers',
        'marker': {
            'color': data.time,
            'size': data.candidate_confidence,
            'showscale': True,
        },
        'text': data.sentiment
    }
]

fig = go.Figure(data=plot_data)
iplot(fig)
```

```python
data_2015 = df[df['year'] == 2015]['sentiment_confidence']
data_2014 = df[df['year'] == 2014]['sentiment_confidence']

trace2 = go.Histogram(
    x = data_2014,
    opacity = 0.75,
    name = '2014',
    marker = dict(color = 'rgba(80, 27, 80, 0.7)'))

trace3 = go.Histogram(
    x = data_2015,
    opacity = 0.75,
    name = '2015',
    marker = dict(color = 'rgba(125, 2, 100, 0.6)'))

data = [trace2, trace3]
layout = go.Layout(barmode = 'overlay',
                   title = 'Frequency of sentiment confidence in 2014 and 2015',
                   xaxis = dict(title = 'sentiment confidence'),
                   yaxis = dict(title = 'count'),
                  )

fig = go.Figure(data = data, layout = layout)
iplot(fig)
```
